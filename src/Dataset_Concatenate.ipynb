## Import's

```python
from datasets import Dataset, concatenate_datasets
import math
import os
```

### Variveis Constantes

```python
DIR = "../"

dataset_path = "data/"
```

### Lendo os Dataset's

```python
dict_datasets : dict[str, Dataset] = dict({})
for arquivo_nome in list(filter(lambda x: x.endswith(".json"), os.listdir(DIR+dataset_path))):
    dict_datasets.update(dict({arquivo_nome: Dataset.from_json(DIR+dataset_path+arquivo_nome)}))
```

```python
dict_datasets
```

### Concatenando os datasets

```python
dataset = concatenate_datasets(list(dict_datasets.values()))
```

```python
dataset
```

### Corrigindo as labels

```python
dataset_aux = dataset.filter(lambda data: len(data["labels"]) == 3)
try:
    assert len(dataset) == len(dataset_aux)
except:
    print("Nem todas as labels possuem 3 classes")
else:
    print("Todas as labels possuem 3 classes")
```

```python
def corrigi_labels(data):
    if len(data["labels"]) < 3:
        data["labels"].insert(1, 0.0)
    return data
```

```python
dataset["labels"][:5]
```

```python
dataset = dataset.map(corrigi_labels)
```

```python
dataset["labels"][:5]
```

```python
dataset_aux = dataset.filter(lambda data: len(data["labels"]) == 3)
try:
    assert len(dataset) == len(dataset_aux)
except:
    print("Nem todas as labels possuem 3 classes")
else:
    print("Todas as labels possuem 3 classes")
```

### Criando agrupamento

#### Removendo o anterior agrupamento

```python
dataset.features
```

```python
dataset = dataset.remove_columns(["group"])
```

```python
dataset.features
```

#### Criando novo agrupamento

```python
lista_grupos = list(range(1, 11)) * math.ceil(len(dataset)/10)
lista_grupos = lista_grupos[:len(dataset)]
```

```python
len(lista_grupos)
```

```python
lista_grupos
```

```python
dataset = dataset.add_column("group", lista_grupos)
```

```python
dataset.features
```

```python
dataset["group"]
```

#### Verificando os grupos

```python
def group_by(dataset: Dataset, coluna: str):
    lista_grupos = []
    for grupo in set(dataset[coluna]):
        lista_grupos.append(dataset.filter(lambda x: x[coluna]==grupo))
    return lista_grupos
```

```python
lista_datasets = group_by(dataset, "group")
```

```python
for index, dataset_part in enumerate(lista_datasets):
    print(str(index+1) + " - " + str(len(dataset_part)))
```
